```python
import pandas as pd

tabela = pd.read_excel("tabela_alunos.xlsx")
display(tabela)
```

# Aqui está em ordem Alfabetica  

```python
tabela.sort_values('Nome', inplace=True)
tabela
```

# Organizei em ordem Crescente o horario 

```python
tabela.sort_values('Hora', inplace=True)
tabela
```

# Entrada e saída

```python
entrada_saida = tabela[["Nome", "Atividade"]].groupby(["Nome"]).sum()
display(entrada_saida)
```

# Saiu e voltou

```python
entrada_saida[entrada_saida["Atividade"] == "EntrouSaiuEntrou"]
```

# Saiu e não voltou :(

```python
entrada_saida.loc[entrada_saida["Atividade"] == "EntrouSaiu"]
```
